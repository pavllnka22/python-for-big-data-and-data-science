15. Порівняйте список та кортеж. Назвіть схожості та відмінності, випадки використання.

Кортежі є незмінними(не дозволяється видаляти або додавати елементи після створення кортежу) та впорядкованими, дозволяються дублікати елементів(бо кожен елемент має індекс), записуються через круглі дужки.

Списки також є впорядкованими, дозволють дублікати(бо кожен елемент має індекс), але на відміну від кортежів списки є змінними, тобто можна додавати або видаляти елементи після створення списку, записуються через квадратні дужки.

Кортежі займають менше памʼяті, ніж списки з таким самим змістом. На кортеж одразу зрозуміло скільки має виділитися памʼяті.

Кортежі використовують, коли треба зберегти дані, які не будуть змінюватися.

# VI. Множини (Set).

17. Створити множину без елементів. Після цого за допомогою методу додайте кілька різних елементів до множини. Чи множини є змінним типом даних?

```python
my_set = set()
my_set.add(1)
my_set.add(2)
print(my_set)
```

Множини є змінним типом даних, можна видаляти та додавати елементи після створення множини, але елементи в множині є незмінними, через це в множину не можна додавати списки, словники та інші множини (бо вони змінні).

18. Створити множину, маючи список my_list = [1, 1, 2, 67, 67, 8, 9]. Пояснити, чому "зникли" деякі елементи.

```python
my_list = [1, 1, 2, 67, 67, 8, 9]
my_set = set(my_list)
print(my_set)
```

У множинах не дозволяються дублікати елементів

19. Створіть дві множини. Продемонстуйте над ними операції: обʼєднання, різниці, пересічі та симетричної різниці. Використовуйте методи, що не змінюють множини, а створюють нові.

```python
my_set1 = {1, 2, 3}
my_set2 = {9, 2, 5}
union_set = my_set1.union(my_set2)
print(union_set)
diff_set = my_set1.difference(my_set2)
print(diff_set)
inter_set = my_set1.intersection(my_set2)
print(inter_set)
sym_diff_set = my_set1.symmetric_difference(my_set2)
print(sym_diff_set)
```

# VII. Словники (Dictionary).

20. Створіть пустий словник. До нього додайте чотири пари елементів такі, щоб їхні ключі були різних типів. Чи може список бути ключем? Чому?

```python
my_dict = {}
my_dict[1] = "first"
my_dict["red"] = "apple"
my_dict[1.12] = 112
my_dict[(1, 2, 3)] = True
print(my_dict)
```

Ні, список не може бути ключем, тому що ключі в словнику мають бути незмінних типів, тому списки, інші словники та множини не можуть бути ключами в словниках. А також ключі не можуть повторюватись.
