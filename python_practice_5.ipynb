6. (11б). Створіть 2 дочірніх класи від 3d-фігура: Циліндр та Куб, які мають атрибути, необхідні для розрахунку об'єму. Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
import math
class ThreeD_figure:

  def get_volume(self):
    return 0

  def __add__(self, other):
    return self.get_volume() + other.get_volume()


class Cylinder(ThreeD_figure):

  def __init__(self, height, radius):
    self.height = height
    self.radius = radius

  def get_volume(self):
    return self.height * self.radius * self.radius * math.pi


class Cube(ThreeD_figure):

  def __init__(self, length):
    self.length = length

  def get_volume(self):
    return pow(self.length, 3)


cube = Cube(2)
cylinder = Cylinder(3, 5)
print(cube.get_volume() + cylinder.get_volume())
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class EWallet:

  def __init__(self, id, balance):
    self.__balance = balance
    self.__id = id

  @property
  def get_wallet_id(self):
        return self.__id

  @property
  def balance(self):
        return self.__balance

  @balance.setter
  def balance(self, value):
      self.__balance = value


  def top_up_balance(self, amount):
    self.__balance += amount


  def withdraw(self, amount):
    if self.balance >= amount:
       self.balance -= amount
    else:
      print("Insufficient funds")


my_wallet = EWallet("56235", 1000)

print(my_wallet.balance)
my_wallet.withdraw(50)
print(my_wallet.balance)
my_wallet.top_up_balance(500)
print(my_wallet.balance)

```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
class Item:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return self.name


class ShoppingCart:


    def __init__(self, items=None):
        if items is None:
            items = []
        else:
            items = list(items)
        self.items = items


    def add_item(self, item):
        if isinstance(item, Item):
            self.items.append(item)
        else:
            raise TypeError('Item must be of type Item')


    def remove_item(self, item):
        if item in self.items:
            self.items.remove(item)
        else:
            print(f'Item {item.name} is not in the shopping cart')


    def remove_all_items(self):
        self.items.clear()


    def show_items(self):
        if len(self.items) == 0:
            print('No items in the shopping cart')
        else:
            print("Items: " + ", ".join(str(item) for item in self.items))


shopping_cart = ShoppingCart()
book = Item('Book')
pen = Item('Pen')
shopping_cart.add_item(book)
shopping_cart.add_item(pen)
shopping_cart.show_items()
shopping_cart.remove_all_items()
shopping_cart.show_items()
```

Ітерабельний обʼєкт - це обʼєкт, по якому можна "пройтись циклом", наприклад: список, кортеж або рядок. Ітератор - це обʼєкт, який виконує ітерацію та повертає наступний елемент ітерабельного обʼєкта.

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class WeekIterator:


    def __init__(self):
        self.days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index < len(self.days):
          day = self.days[self.index]
          self.index += 1
          return day
        else:
          raise StopIteration


week = WeekIterator()
for day in week:
    print(day)
```
